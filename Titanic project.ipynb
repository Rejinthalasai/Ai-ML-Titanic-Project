```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
```

```python
#We have to know types of embarked in the columns
Embarked = df['Embarked'].unique()
for Embarkeds in Embarked:
    print("*",Embarkeds)
```

```python
Sex = df['Sex'].unique()
for i in Sex:
    print("*",i)
```

```python

df['Embarked'] = df['Embarked'].map( {'Q': 0,'S':1,'C':2}).astype(int)
df['Sex'] = df['Sex'].map( {'female': 1,'male':0}).astype(int)
```

```python
## Here Age and Fare are in FLoat Type we have to change them into Integer
df['Age'] = df['Age'].astype(int)
df['Fare'] = df['Fare'].astype(int)
```

```python

data = df.drop(['PassengerId','Name','Cabin','Ticket'], axis =1, inplace=True)
```

## Spliting Data

```python
Train = df.drop(['Survived'], axis=1)
Test = df.iloc[:,1]
x_train, x_test, y_train, y_test = train_test_split(Train, Test, test_size = 0.2, random_state = 1)
```

## 5.Model Building

```python
#this specifies the algorithm used fro optimaization.
LR = LogisticRegression(solver='liblinear', max_iter=200)
#fit: This trains the logistic regression model using the training data (x_train for features and y_train for target labels). 
LR.fit(x_train, y_train)
y_pred = LR.predict(x_test)
LRA = accuracy_score(y_pred,y_test)
print('Accuracy: {:.2f}%'.format(LRA*100))
```

```python
import joblib

# Assuming LR is your trained LogisticRegression model
joblib.dump(LR, 'logistic_regression_model.pkl')
```
